```python
import numpy as np
import pandas as pd
eps = np.finfo(float).eps
from numpy import log2 as log
```

```python
df = pd.read_csv('diseases.csv', index_col=0, parse_dates=True)
df
```

```python

entropy_node = 0
values = df.Result.unique()
for value in values:
    fraction = df.Result.value_counts()[value]/len(df.Result)
    entropy_node += -fraction*np.log2(fraction)

print(f'Values: {values}')
print(f'entropy_node: {entropy_node}')
```

```python
def ent(df,attribute):
    target_variables = df.Result.unique()
    variables = df[attribute].unique()

    entropy_attribute = 0
    for variable in variables:
        entropy_each_feature = 0
        for target_variable in target_variables:
            num = len(df[attribute][df[attribute]==variable][df.Result ==target_variable])
            den = len(df[attribute][df[attribute]==variable])
            fraction = num/(den+eps)
            entropy_each_feature += -fraction*log(fraction+eps)
        fraction2 = den/len(df)
        entropy_attribute += -fraction2*entropy_each_feature

    return(abs(entropy_attribute))
```

```python
a_entropy = {k:ent(df,k) for k in df.keys()[:-1]}
a_entropy
```

```python
def find_entropy(df):
    Class = df.keys()[-1]
    entropy = 0
    values = df[Class].unique()
    for value in values:
        fraction = df[Class].value_counts()[value]/len(df[Class])
        entropy += -fraction*np.log2(fraction)
    return entropy


def find_entropy_attribute(df,attribute):
  Class = df.keys()[-1]
  target_variables = df[Class].unique()
  variables = df[attribute].unique()
  entropy2 = 0
  for variable in variables:
      entropy = 0
      for target_variable in target_variables:
          num = len(df[attribute][df[attribute]==variable][df[Class] ==target_variable])
          den = len(df[attribute][df[attribute]==variable])
          fraction = num/(den+eps)
          entropy += -fraction*log(fraction+eps)
      fraction2 = den/len(df)
      entropy2 += -fraction2*entropy
  return abs(entropy2)


def find_winner(df):
    Entropy_att = []
    IG = []
    for key in df.keys()[:-1]:
        IG.append(find_entropy(df)-find_entropy_attribute(df,key))
    return df.keys()[:-1][np.argmax(IG)]


def get_subtable(df, node,value):
  return df[df[node] == value].reset_index(drop=True)


def buildTree(df,tree=None):
    Class = df.keys()[-1]
    node = find_winner(df)
    attValue = np.unique(df[node])

    if tree is None:
        tree={}
        tree[node] = {}
    for value in attValue:

        subtable = get_subtable(df,node,value)
        clValue,counts = np.unique(subtable[Class],return_counts=True)

        if len(counts)==1:
            tree[node][value] = clValue[0]
        else:
            tree[node][value] = buildTree(subtable)

    return tree

t  = buildTree(df)
import pprint
pprint.pprint(t)
```

```python
import pydot
import uuid

def generate_unique_node():
    return str(uuid.uuid1())

def create_node(graph, label, shape='oval'):
    node = pydot.Node(generate_unique_node(), label=label, shape=shape)
    graph.add_node(node)
    return node

def create_edge(graph, node_parent, node_child, label):
    link = pydot.Edge(node_parent, node_child, label=label)
    graph.add_edge(link)
    return link

def walk_tree(graph, dictionary, prev_node=None):

    for parent, child in dictionary.items():
        # root
        if not prev_node:
            root = create_node(graph, parent)
            walk_tree(graph, child, root)
            continue

        # node
        if isinstance(child, dict):
            for p, c in child.items():
                n = create_node(graph, p)
                create_edge(graph, prev_node, n, str(parent))
                walk_tree(graph, c, n)

        # leaf
        else:
            leaf = create_node(graph, str(child), shape='box')
            create_edge(graph, prev_node, leaf, str(parent))


tree={'Debility': {'no': {'Fever': {'36-38': 'no',
                               '38-40': {'Age': {'1-18': 'no',
                                                 '18-35': 'no',
                                                 '35-60': 'no',
                                                 '>60': {'Chronic_pain': {'Occasionally': 'no',
                                                                          'always': {'Gender': {'female': 'yes',
                                                                                                'male': 'no'}},
                                                                          'often': {'Gender': {'female': {'Digestive_issues': {'no': 'yes',
                                                                                                                               'yes': 'no'}},
                                                                                               'male': 'no'}},
                                                                          'sometimes': 'no'}}}},
                               '<36': {'Age': {'1-18': 'no',
                                               '18-35': 'no',
                                               '35-60': 'no',
                                               '>60': {'Chronic_pain': {'Occasionally': {'Digestive_issues': {'no': 'no',
                                                                                                              'yes': 'yes'}},
                                                                        'often': {'Gender': {'female': {'Digestive_issues': {'no': 'yes',
                                                                                                                             'yes': 'no'}},
                                                                                             'male': 'yes'}},
                                                                        'sometimes': 'no'}}}},
                               '>40': {'Chronic_pain': {'Occasionally': {'Age': {'1-18': 'no',
                                                                                 '18-35': {'Gender': {'female': 'yes',
                                                                                                      'male': 'no'}},
                                                                                 '35-60': {'Gender': {'female': 'no',
                                                                                                      'male': {'Digestive_issues': {'no': 'no',
                                                                                                                                    'yes': 'yes'}}}},
                                                                                 '>60': {'Gender': {'female': 'yes',
                                                                                                    'male': 'no'}}}},
                                                        'always': {'Age': {'1-18': 'yes',
                                                                           '18-35': {'Gender': {'female': 'yes',
                                                                                                'male': 'no'}},
                                                                           '35-60': 'yes',
                                                                           '>60': 'yes'}},
                                                        'often': {'Age': {'1-18': 'no',
                                                                          '18-35': 'no',
                                                                          '35-60': {'Gender': {'female': 'no',
                                                                                               'male': 'yes'}},
                                                                          '>60': {'Gender': {'female': 'yes',
                                                                                             'male': {'Digestive_issues': {'no': 'no',
                                                                                                                           'yes': 'yes'}}}}}},
                                                        'sometimes': {'Digestive_issues': {'no': 'no',
                                                                                           'yes': {'Age': {'1-18': 'no',
                                                                                                           '18-35': 'yes',
                                                                                                           '35-60': 'yes',
                                                                                                           '>60': {'Gender': {'female': 'yes',
                                                                                                                              'male': 'no'}}}}}}}}}},
              'yes': {'Fever': {'36-38': {'Chronic_pain': {'Occasionally': 'no',
                                                           'always': {'Digestive_issues': {'no': {'Age': {'1-18': 'no',
                                                                                                          '18-35': 'no',
                                                                                                          '35-60': 'no',
                                                                                                          '>60': {'Gender': {'female': 'no',
                                                                                                                             'male': 'yes'}}}},
                                                                                           'yes': {'Age': {'1-18': 'yes',
                                                                                                           '18-35': {'Gender': {'female': 'yes',
                                                                                                                                'male': 'no'}},
                                                                                                           '35-60': 'yes',
                                                                                                           '>60': 'yes'}}}},
                                                           'often': {'Age': {'1-18': 'no',
                                                                             '18-35': 'no',
                                                                             '35-60': 'no',
                                                                             '>60': {'Gender': {'female': {'Digestive_issues': {'no': 'no',
                                                                                                                                'yes': 'yes'}},
                                                                                                'male': 'no'}}}},
                                                           'sometimes': {'Age': {'1-18': 'no',
                                                                                 '18-35': 'no',
                                                                                 '35-60': 'no',
                                                                                 '>60': {'Gender': {'female': 'yes',
                                                                                                    'male': 'no'}}}}}},
                                '38-40': {'Chronic_pain': {'Occasionally': {'Age': {'1-18': 'no',
                                                                                    '18-35': 'no',
                                                                                    '35-60': 'no',
                                                                                    '>60': {'Digestive_issues': {'no': 'no',
                                                                                                                 'yes': 'yes'}}}},
                                                           'always': 'yes',
                                                           'often': {'Digestive_issues': {'no': {'Age': {'1-18': 'no',
                                                                                                         '18-35': 'no',
                                                                                                         '35-60': {'Gender': {'female': 'yes',
                                                                                                                              'male': 'no'}},
                                                                                                         '>60': 'yes'}},
                                                                                          'yes': 'yes'}},
                                                           'sometimes': {'Age': {'1-18': 'no',
                                                                                 '18-35': 'no',
                                                                                 '35-60': 'no',
                                                                                 '>60': {'Gender': {'female': {'Digestive_issues': {'no': 'no',
                                                                                                                                    'yes': 'yes'}},
                                                                                                    'male': 'yes'}}}}}},
                                '<36': {'Age': {'1-18': {'Chronic_pain': {'Occasionally': 'no',
                                                                          'always': {'Digestive_issues': {'no': 'no',
                                                                                                          'yes': 'yes'}},
                                                                          'often': 'no',
                                                                          'sometimes': 'no'}},
                                                '18-35': {'Digestive_issues': {'no': 'no',
                                                                               'yes': {'Chronic_pain': {'Occasionally': 'no',
                                                                                                        'always': 'yes',
                                                                                                        'often': 'yes',
                                                                                                        'sometimes': 'yes'}}}},
                                                '35-60': {'Digestive_issues': {'no': {'Chronic_pain': {'Occasionally': 'no',
                                                                                                       'always': 'no',
                                                                                                       'often': 'yes',
                                                                                                       'sometimes': 'no'}},
                                                                               'yes': {'Chronic_pain': {'Occasionally': {'Gender': {'female': 'no',
                                                                                                                                    'male': 'yes'}},
                                                                                                        'always': 'yes',
                                                                                                        'often': 'yes',
                                                                                                        'sometimes': 'yes'}}}},
                                                '>60': {'Gender': {'female': 'yes',
                                                                   'male': {'Digestive_issues': {'no': {'Chronic_pain': {'Occasionally': 'yes',
                                                                                                                         'always': 'yes',
                                                                                                                         'often': 'no',
                                                                                                                         'sometimes': 'no'}},
                                                                                                 'yes': 'yes'}}}}}},
                                '>40': {'Chronic_pain': {'Occasionally': {'Digestive_issues': {'no': {'Age': {'1-18': 'no',
                                                                                                              '18-35': 'no',
                                                                                                              '35-60': 'yes',
                                                                                                              '>60': {'Gender': {'female': 'no',
                                                                                                                                 'male': 'yes'}}}},
                                                                                               'yes': 'yes'}},
                                                         'always': 'yes',
                                                         'often': 'yes',
                                                         'sometimes': {'Age': {'1-18': {'Digestive_issues': {'no': 'no',
                                                                                                             'yes': 'yes'}},
                                                                               '18-35': 'yes',
                                                                               '35-60': {'Gender': {'female': 'yes',
                                                                                                    'male': {'Digestive_issues': {'no': 'no',
                                                                                                                                  'yes': 'yes'}}}},
                                                                               '>60': 'yes'}}}}}}}}

graph = pydot.Dot(graph_type='digraph')
walk_tree(graph, tree)

graph.write_png("DecisionTree.png")
```
